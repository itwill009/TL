```python
state_party_vote = primary.groupby(['state','party'])['votes'].sum()
```

```python
state_vote = primary.groupby('state')['votes'].sum()
```

```python
vote_pot = state_party_vote / state_vote
```

```python
vote_pot.head()
```

```python
vote_pot.unstack().plot(kind='barh',stacked=True)
```

```python
func = lambda age_df : age_df.sort_values('votes',ascending=False).iloc[0]
```

```python
primary.head()
```

```python
winners = primary.groupby('fips').agg(func)
```

```python
winners.head()
```

```python
counties.columns
```

```python
counties.RHI825214.head()
```

```python
winners_county = pd.merge(winners,counties[['fips','RHI825214']],left_index=True,right_on='fips',how='left')
```

```python
winners_county.head()
```

```python
winner_race = winners_county.rename(columns={'RHI825214':'white_pcts'})
```

```python
winner_race.head()
```

```python
winner_race_pots = winner_race.groupby(['party','candidate'])['white_pcts'].mean()
```

```python
winner_race_pots.head()
```

```python
ax = winner_race_pots.plot(kind='barh')
ax.set_xlim([50,100])
```
